# Hospital Readmissions Data Analysis and Recommendations for Reduction

### Background
In October 2012, the US government's Center for Medicare and Medicaid Services (CMS) began reducing Medicare payments for Inpatient Prospective Payment System hospitals with excess readmissions. Excess readmissions are measured by a ratio, by dividing a hospital’s number of “predicted” 30-day readmissions for heart attack, heart failure, and pneumonia by the number that would be “expected,” based on an average hospital with similar patients. A ratio greater than 1 indicates excess readmissions.

Since there is no linear regression, the observation that "overall, rate of readmissions is trending down with increasing number of discharges" is unfounded. Moreover since there are different measures in the data, there could be differences for admission rates for different measures, which may not be shown by the initial observation/plots. Also while the plot suggests a linear model is a poor explanation of variance. (We can measure this by calculating r^2)

```python
x = [a for a in clean_hospital_read_df['Number of Discharges'][81:-3]]
y = list(clean_hospital_read_df['Excess Readmission Ratio'][81:-3])
data = pd.DataFrame([x,y], index=['Number of Discharges', 'Excess Readmission Ratio']).transpose()
f = sns.lmplot(x='Number of Discharges', y='Excess Readmission Ratio', data=data)
```

```python
slope, intercept, rvalue, pvalue, stderr = stats.linregress(x, y)
print("Slope:"+str(slope))
print("Intercept:"+str(intercept))
print("R-Squared:"+str(rvalue**2))
print("P-Value:"+str(pvalue))
print("stderr:"+str(stderr))
```

Value of r-squared is very small so the variance in the excess readmission ratio is not explained by the number of discharges.  Conversely because the p-value is <.05 we can reject the null hypothesis that the best fit regression line has no slope.

```python
set(df['Measure Name'])
x, y , hue = 'Number of Discharges', 'Excess Readmission Ratio', 'Measure Name'
f = sns.lmplot(x=x, y=y, hue=hue, data=df[[x,y,hue]])
```
