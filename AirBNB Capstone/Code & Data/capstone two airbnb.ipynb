```python
import pandas as pd
import numpy as np
import seaborn as sns
%cd C:\Users\Shing Han\Documents\Springboard\capstone 2
listings = pd.read_csv('listings.csv')
print(listings.shape)
listings.transpose()
listings['price'] = listings.price.str.replace("\$|,",'').astype(float)
#Clean Data: we want to remove all non numeric values for the time being
listings = listings.select_dtypes(include=[np.number])
listings.transpose()
```

```python
#We now want to drop some more of the columns (NaN ones)
drop_list = ['id','scrape_id','host_id',
             'thumbnail_url','medium_url','xl_picture_url',
             'host_acceptance_rate','neighbourhood_group_cleansed','square_feet']
listings = listings.drop(drop_list, axis=1)
listings = listings.replace(np.NaN, 0)
listings.transpose()
```

```python
#Import necessary packages for machine learning
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error
from scipy.spatial import distance
#Univariate Model
test_review_rating = 80
first_review_value = listings.loc[0,'review_scores_rating']
#Calculate first euclidean distance
first_distance = np.abs(first_review_value - test_review_rating)
print(first_distance)
listings['distance'] = np.abs(listings.review_scores_rating - test_review_rating)
listings.distance.value_counts().sort_index()
```

```python
#Use random_state parameter which gives a reproducible random order 
#so you can get the same results
listings = listings.sample(frac=1,random_state=0)
listings = listings.sort_values('distance')
```

```python
mean_price = listings.price.iloc[:5].mean()
mean_price
#We can now assume an 80 rating will net a mean_price of ~ $111.60
```

We will now evaluate the model by splitting the data into
train and test sets (70%/30%), using sklearn train_test_split.  We will then use training sets to predict the price for rows in test set to compare the predicted values with actual price values to test for accuracy.


```python
listings.drop('distance',axis=1)
train_df = listings.copy().iloc[:3603]
test_df = listings.copy().loc[3603:]
```

```python
#Create a function for the univariate function
def prediction(new_listing, feature_column):
    temp_df = train_df
    temp_df['distance'] = np.abs(listings[feature_column]-new_listing)
    temp_df = temp_df.sort_values('distance')
    knn_5 = temp_df.price.iloc[:5]
    price_prediction = knn_5.mean()
    return(price_prediction)
```

```python
#Predict price of review_score_ratings
test_df['price_prediction']=test_df.review_scores_rating.apply(prediction,feature_column='review_scores_rating')
```

```python
test_df['squared_error']=(test_df['price_prediction']-test_df['price'])**(2)
mse = test_df['squared_error'].mean()
rmse = mse ** (1/2)
rmse
#This is a very high RMSE...
```

```python
#We are going to normalize the data
listings_norm = (listings-listings.mean())/(listings.max()-listings.min())
listings_norm.transpose()
```

```python
norm_train_df = listings_norm.copy().iloc[:3603]
norm_test_df = listings_norm.copy().loc[3603:]
```

```python
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error
knn = KNeighborsRegressor(algorithm='brute')
cols1 = ['accommodates','bedrooms','bathrooms','beds']
knn.fit(norm_train_df[cols1],norm_train_df['price'])
model1_predict = knn.predict(test_df[cols1])
model1_mse = mean_squared_error(norm_test_df['price'], model1_predict)
model1_rmse = model1_mse ** (1/2)
model1_rmse
```

```python
cols2 = ['accommodates','bedrooms','bathrooms']
knn.fit(norm_train_df[cols2],norm_train_df['price'])
model2_predict = knn.predict(test_df[cols2])
model2_mse = mean_squared_error(norm_test_df['price'], model2_predict)
model2_rmse = model2_mse ** (1/2)
model2_rmse
```

```python
cols3 = ['accommodates','bedrooms','bathrooms','beds','review_scores_rating']
knn.fit(norm_train_df[cols3],norm_train_df['price'])
model3_predict = knn.predict(test_df[cols3])
model3_mse = mean_squared_error(norm_test_df['price'], model3_predict)
model3_rmse = model3_mse ** (1/2)
model3_rmse
```
